```python
category_sales = df.groupby('Category')['Total_Sale'].sum().sort_values(ascending=False)
category_sales
```

```python
top_medicines = df.groupby('Medicine_Name')['Total_Sale'].sum().sort_values(ascending=False).head(10)
top_medicines
```

```python
gender_sales = df.groupby('Gender')['Total_Sale'].sum()
gender_sales
```

```python
age_sales = df.groupby('Age_Group')['Total_Sale'].sum().sort_values(ascending=False)
age_sales
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
monthly_sales = df.groupby(df['Date'].dt.to_period('M'))['Total_Sale'].sum()
monthly_sales
```

```python
df['Expiry_Date'] = pd.to_datetime(df['Expiry_Date'])
```

```python
expiring = df[
    df['Expiry_Date'] <= pd.Timestamp.today() + pd.DateOffset(months=6)
]

expiring[['Medicine_Name', 'Expiry_Date']].head(20)
```

```python
low_stock = df[['Medicine_Name', 'Stock_Available']].sort_values('Stock_Available').head(10)
low_stock
```

```python
payment_sales = df.groupby('Payment_Mode')['Total_Sale'].sum().sort_values(ascending=False)
payment_sales
```

```python
location_sales = df.groupby('Store_Location')['Total_Sale'].sum().sort_values(ascending=False)
location_sales
```
